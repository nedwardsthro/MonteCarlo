```python
from montecarlo import *
import numpy as np
import pandas as pd
```

## Die Class

```python
Demo_Die = Die([1,2,3,4])
Demo_Die.show()
```

```python
Demo_Die.change_weight(3,3)
Demo_Die.show()
```

```python
Demo_Die.roll(10)
```

## Game Class

```python
Demo_Game = Game([Demo_Die, Demo_Die, Demo_Die])
```

```python
Demo_Game.play(100)
Demo_Game.show("wide")
```

```python
Demo_Game.show("narrow")
```

## Analyzer Class

```python
Demo_Analyzer = Analyzer(Demo_Game)
Demo_Analyzer.face_counts_per_roll()
Demo_Analyzer.face_counts
```

```python
Demo_Analyzer.combo()
Demo_Analyzer.combo
```
